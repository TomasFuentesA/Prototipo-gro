```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('Documents\XOR_circuit3_I_1.csv')

df = df[:1300]
variables_w = [' W0', ' W1', ' Rep']
colores = ['magenta', 'aquamarine', 'orange']

# Normalizar los datos
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df[variables_w]), columns=variables_w)

# Crear un grfico para las tres variables normalizadas en el mismo eje
plt.figure()
for variable, color in zip(variables_w, colores):
    plt.plot(df['Time'], df_normalized[variable], label=f'{variable} (Normalizada)', color=color)

plt.xlabel('Tiempo')
plt.ylabel('Valores Normalizados')
plt.title('Variables normalizadas vs. Tiempo')
plt.legend()
plt.show()
```

```python
df = df[:1500]

# Mostrar todos los grficos juntos
plt.show()

# Crear un grfico para las tres variables normalizadas en el mismo eje
plt.figure()
for variable, color in zip(variables_w, colores):
    plt.plot(df['Time'], df[variable], label=f'{variable}', color=color)

plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Seales vs. Tiempo')
plt.legend()
plt.show()
```
